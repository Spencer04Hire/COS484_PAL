```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Results.csv")
languages = df["Language"].unique()
datasets = np.array(df["Dataset"].unique())

all_results = {}
for language in languages:
    all_results[language] = np.array(df[df["Language"] == language]["Accuracy"])

# print(all_results)

plt.figure(figsize = (32, 16))
java = all_results["Java"]
bar = np.arange(len(java))
for key, value in all_results.items():
    plt.bar(bar, value, width = 0.1, label = key)
    # break
    bar = [x + 0.1 for x in bar]

plt.xlabel('Dataset', fontsize = 15) 
plt.ylabel('Accuracy', fontsize = 15)
plt.title("Accuracy of Prompting Strategies on Different Datasets")
plt.xticks([r + 0.25 for r in range(len(java))], 
        datasets)
plt.legend()
plt.show()
```
