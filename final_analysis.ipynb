```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mathematical_results.csv")
languages = ['Direct', 'CoT', 'C++', 'Java', 'OCaml', 'Python']
math_datasets = df["Dataset"].unique()

math_results = {}
for language in languages:
    math_results[language] = np.array(df[df["Language"] == language]["Accuracy"])

plt.figure(figsize = (40, 16))
bar = np.arange(len(math_datasets))
for key, value in math_results.items():
    plt.bar(bar, value, width = 0.1, label = key)
    bar = [x + 0.1 for x in bar]

plt.xlabel('Dataset', fontsize = 24) 
plt.ylabel('Accuracy', fontsize = 24)
plt.title("Accuracy of Different Programming Languages on Mathematical Datasets", fontsize = 40)
plt.xticks([r + 0.25 for r in range(len(math_datasets))], 
        math_datasets, fontsize = 20)
plt.yticks(fontsize=20)
plt.legend(fontsize = 16)
plt.show()
```
