```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Results.csv")
languages = df["Language"].unique()
math_datasets = np.array(df["Dataset"].unique())[:7]

all_results = {}
for language in languages:
    all_results[language] = np.array(df[df["Language"] == language]["Accuracy"])

# print(all_results)

plt.figure(figsize = (40, 16))
java = all_results["Java"][:7]
bar = np.arange(len(java))
for key, value in all_results.items():
    plt.bar(bar, value[:7], width = 0.1, label = key)
    # break
    bar = [x + 0.1 for x in bar]

plt.xlabel('Dataset', fontsize = 14) 
plt.ylabel('Accuracy', fontsize = 14)
plt.title("Accuracy of Different Programming Languages on Mathematical Datasets", fontsize = 40)
plt.xticks([r + 0.25 for r in range(len(java))], 
        math_datasets, fontsize = 12)
plt.legend(fontsize = 14)
plt.show()
```

```python
df = pd.read_csv("Results.csv")
languages = df["Language"].unique()
reasoning_datasets = np.array(df["Dataset"].unique())[7:]

all_results = {}
for language in languages:
    all_results[language] = np.array(df[df["Language"] == language]["Accuracy"])


plt.figure(figsize = (40, 16))
java = all_results["Java"][7:]
bar = np.arange(len(java))
for key, value in all_results.items():
    plt.bar(bar, value[7:], width = 0.1, label = key)
    bar = [x + 0.1 for x in bar]

plt.xlabel('Dataset', fontsize = 14) 
plt.ylabel('Accuracy', fontsize = 14)
plt.title("Accuracy of Different Programming Languages on Reasoning Datasets", fontsize = 40)
plt.xticks([r + 0.25 for r in range(len(java))], 
        reasoning_datasets, fontsize = 12)
plt.legend(fontsize = 14)
plt.show()
```
